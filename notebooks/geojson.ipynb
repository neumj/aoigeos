```python
import pandas as pd
import json
```

```python
def read_json(file_path):
    with open(file_path) as json_data:
        d = json.load(json_data)
    return d

def write_json(data_dict,file_path):
    with open(file_path, 'w') as fp:
        json.dump(data_dict, fp, indent=4)
```

```python
poly_str = df.iloc[0]['polygon']
def wkt_to_geojson(polygon_str):
    raw_coords_str = polygon_str.split('((')[1].split('))')[0].strip()
    clean_coords = [[float(a.strip().replace(' ',',').split(',')[0]), \
                 float(a.strip().replace(' ',',').split(',')[1])] \
                for a in raw_coords_str.split(',')]
    return clean_coords
```
